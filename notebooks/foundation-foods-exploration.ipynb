## 1. Load data

````python
foundation_foods = pd.read_csv('../data/cleaned_foundation_food.csv')
food_nutrients = pd.read_csv('../data/cleaned_food_nutrient.csv')
nutrients = pd.read_csv('../data/cleaned_nutrient.csv')
````

## 2. Restrict nutrients to Foundation Foods

````python
# restrict food_nutrients to foundation food
foundation_food_nutrients = food_nutrients[food_nutrients.apply(lambda x: x['fdc_id'] in foundation_foods['fdc_id'].values,axis=1)]

# pivot dataframe
pivot = foundation_food_nutrients.pivot(index='fdc_id', columns='nutrient_id', values='amount').reset_index()

# identify nutrients in foundation foods
nutrients_in_foundation_foods=foundation_food_nutrients['nutrient_id'].unique()
````

## 3. Group nutrients into categories

````python
# convert nutrients to dict
nutrients.index=nutrients['id']
nutrients_dict=nutrients['name'].to_dict()
nutrients_dict[2066]='' # resolves KeyError
````

````python
nutrient_names=list(map(lambda x: nutrients_dict[x].lower(),nutrients_in_foundation_foods))
````
